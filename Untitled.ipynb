```python
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
```

```python
text = "Food was excellent and delivery was fast"

sia.polarity_scores(text)
```

```python
def get_sentiment(text):

    score = sia.polarity_scores(str(text))

    compound = score['compound']

    if compound >= 0.05:
        return "Positive"

    elif compound <= -0.05:
        return "Negative"

    else:
        return "Neutral"
```

```python
sample_df = df.head(1000).copy()

sample_df['Sentiment'] = sample_df['reviews_list'].apply(get_sentiment)

sample_df['Sentiment'].value_counts()
```

```python
df_small = df[['reviews_list']].copy()

df_small['Sentiment'] = df_small['reviews_list'].astype(str).str[:500].apply(get_sentiment)
```

```python
df_sample = df.sample(10000, random_state=42)
```

```python
df_sample['Sentiment'] = df_sample['reviews_list'].astype(str).str[:500].apply(get_sentiment)
```

```python
df['Sentiment'] = (
    df['reviews_list']
    .astype(str)
    .str[:500]      # first 500 chars only
    .apply(get_sentiment)
)
```

```python
import matplotlib.pyplot as plt

sentiment_counts = df['Sentiment'].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct='%1.1f%%'
)

plt.title("Review Sentiment Distribution")

plt.show()
```
